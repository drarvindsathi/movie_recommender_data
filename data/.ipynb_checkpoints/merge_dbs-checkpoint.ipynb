```python
import pandas as pd
import re
import numpy as np
import json
```

```python
df_movies = pd.read_csv("movies.csv")
df_movies
```

```python
def extract_movie_title(x):
    split_list = re.split('[()]', x)
    movie_title = split_list[0]
    movie_year = split_list[len(split_list)-2]
    return movie_title

def extract_movie_year(x):
    split_list = re.split('[()]', x)
    movie_title = split_list[0]
    movie_year = split_list[len(split_list)-2]
    return movie_year

def extract_movie_genres(x):
    genre_list = re.split('[|]', x)
    return genre_list
```

```python
test_string = "Twelve Monkeys (a.k.a. 12 Monkeys) (1995)"
movie_title_1 = extract_movie_title(test_string)
print(movie_title_1)
test_genre = "Adventure|Children|Fantasy"
genre_list_1 = extract_movie_genres(test_genre)
print(genre_list_1)
```

```python
df_movies['movie_title']=df_movies['title'].apply(lambda x: extract_movie_title(x))
df_movies['movie_year']=df_movies['title'].apply(lambda x: extract_movie_year(x))
df_movies['movie_genre']=df_movies['genres'].apply(lambda x: extract_movie_genres(x))
df_movies
```

```python
genre_1 = "Drama"
df_movie_in_genre = df_movies[df_movies['movie_genre'].apply(lambda x: genre_1 in x)]
df_movie_in_genre.shape[0]
```

```python
df_links = pd.read_csv("links.csv")
df_links
```

```python
df_movie_abt_1 = pd.merge(df_movies, df_links, 
                                    on=['movieId'], 
                                    suffixes=('', '_DROP'), 
                                    how='outer').filter(regex='^(?!.*_DROP)')
df_movie_abt_1
```

```python
df_tmdb = pd.read_csv("tmdb_5000_movies.csv")
df_tmdb
```

```python
df_movie_abt_2 = pd.merge(df_movie_abt_1, df_tmdb, 
                                    left_on='tmdbId', 
                                    right_on='id',
                                    suffixes=('', '_TMDB'), 
                                    how='outer')
df_movie_abt_2
```

```python
df_imdb = pd.read_csv("imdb_attributes.csv")
df_imdb
```

```python
df_movie_abt_3 = pd.merge(df_movie_abt_2, df_imdb, 
                                    left_on='imdbId', 
                                    right_on='tid',
                                    suffixes=('', '_IMDB'), 
                                    how='outer')
df_movie_abt_3
```

```python
df_movie_abt_3.to_csv("movie_abt.csv")
```

```python
movie_list = []
movie_json_list = json.loads(df_movie_abt_3.to_json(orient='records'))
for movie_json in movie_json_list:
    if movie_json['movie_title'] not in movie_list:
        movie_list.append(movie_json['movie_title'])
    if movie_json['title_TMDB'] not in movie_list:
        movie_list.append(movie_json['title_TMDB'])
    if movie_json['title_IMDB'] not in movie_list:
        movie_list.append(movie_json['title_IMDB'])
len(movie_list)
```

```python
df_movie_list = pd.DataFrame.from_dict(movie_list, orient='columns')
df_movie_list
```

```python
df_movie_list.to_csv('movie_list.csv')
```
