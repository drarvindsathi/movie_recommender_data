```python
import pandas as pd
import json
```

```python

df_movie_abt = pd.read_csv("https://raw.githubusercontent.com/drarvindsathi/movie_recommender_data/main/data/movie_abt.csv")
df_movie_abt
```

```python
movie_json_list = json.loads(df_movie_abt.to_json(orient='records'))
```

```python
genre_1 = "Drama"
movie_in_genre = []
for movie_json in movie_json_list:
    if movie_json['movie_genre']:
        if genre_1 in movie_json['movie_genre']:
            movie_in_genre.append(movie_json)
len(movie_in_genre)
```

```python
def query_builder(dict):
    intent = dict["intent"]
    entity_type = dict["entity_type"]
    entity_value = dict["entity_value"]
    df_movie_abt = pd.read_csv("https://raw.githubusercontent.com/drarvindsathi/movie_recommender_data/main/data/movie_abt.csv")
    movie_json_list = json.loads(df_movie_abt.to_json(orient="records"))
    if intent == 'count_movies_by_genre':
        if entity_type == "genre":
            genre_to_count = entity_value
            movie_in_genre = []
            for movie_json in movie_json_list:
                if movie_json['movie_genre']:
                    if genre_to_count in movie_json["movie_genre"]:
                        movie_in_genre.append(movie_json)
            return {'message': len(movie_in_genre)}
```

```python
data_1 = {"intent": "count_movies_by_genre",
          "entity_type": "genre",
          "entity_value": "Drama"}
query_builder(data_1)
```
